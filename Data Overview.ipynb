```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('bmh')
```

```python
df = pd.read_csv('data_full_01_03_2022.csv')
df['task_completed_sec'] = df['assignment_submitted_time'] - df['assignment_started_time']
df
```

```python
plt.figure(figsize=(12, 6))
plt.title('Time of task completion')
plt.hist(df['task_completed_sec'], bins=100, density=True)
plt.show()
```

# Question 1. 
RU: Можно ли назвать конфликт между Украиной и РФ войной?

EN: Can the conflict between Ukraine and the Russian Federation be called a war?

Answers:
 1. Да / Yes (yes)
 2. Нет / No (no)
 3. Затрудняюсь ответить / Difficult to answer (uncertain)

Column: is_war

```python
plt.figure(figsize=(14, 6))
plt.title('Can the conflict between Ukraine and the Russian Federation be called a war?')
ax = sns.histplot(df['is_war'], stat="percent")
ax.bar_label(ax.containers[0])
plt.xlabel('Answer')
plt.show()
```
